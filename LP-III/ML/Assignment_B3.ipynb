## Preprocessing 

```python
# Remove useless columns
df.drop(['CustomerId','Surname'], axis='columns', inplace=True)
```

```python
df['Gender'].replace({'Female':1,'Male':0}, inplace=True)
```

```python
df.sample(5)
```

```python
df = pd.get_dummies(data=df, columns=['Geography']) # pd.get_dummies() is used for one-hot encoding categorical variables
# The result of this operation is a DataFrame where the original 'Geography' column has been replaced with multiple columns,
# each representing a unique category in the original 'Geography' column. Each row has a 1 in the column corresponding to 
# its 'Geography' category and 0 in the other columns.
```

```python
df.head()
```

```python
df.columns
```

## Scaling of Data 

```python
columns_to_scale = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df.head()
```

```python
x = df.drop('Exited', axis='columns')
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train = np.asarray(x_train).astype(np.int32)
x_test = np.asarray(x_test).astype(np.int32)
y_train = np.asarray(y_train).astype(np.int32)
y_test = np.asarray(y_test).astype(np.int32)
```

## Building Neutral Network 

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(13,), activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred = np.asarray(y_pred).astype(np.int32)
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report: ")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy: ",accuracy_score(y_test,y_pred))
print("Precision: ",precision_score(y_test,y_pred))
print("Recall: ",recall_score(y_test,y_pred))
print("F1: ",f1_score(y_test,y_pred))
```
