## Using K-Nearest Neighbors and Support Vector Machine for Classification 

```python
x = df.iloc[:,1:3001]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

## K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
y_pred_knn
```

```python
cm = confusion_matrix(y_test, y_pred_knn)
cm
```

```python
print("Accuracy: ",accuracy_score(y_test, y_pred_knn))
print("Precision: ",precision_score(y_test, y_pred_knn))
print("Recall: ",recall_score(y_test, y_pred_knn))
print("F1: ",f1_score(y_test, y_pred_knn))
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear',random_state=42)
svc.fit(x_train,y_train)
```

```python
y_pred_svc = svc.predict(x_test)
y_pred_svc
```

```python
cm = confusion_matrix(y_test, y_pred_svc)
cm
```

```python
print("Accuracy: ",accuracy_score(y_test, y_pred_svc))
print("Precision: ",precision_score(y_test, y_pred_svc))
print("Recall: ",recall_score(y_test, y_pred_svc))
print("F1: ",f1_score(y_test, y_pred_svc))
```

## Conclusion 
Upon analysis, SVM model performs better in terms of both Accuracy and F1 scores than KNN model.
