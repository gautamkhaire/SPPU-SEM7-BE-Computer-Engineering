Implement Gradient Descent Algorithm to find the local minima of a function.<br>
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.


## Gradient Descent Algorithm 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x+3)**2
```

```python
def gradient_descent(starting_point, learning_rate, iterations):
    x_values = []
    y_values = []
    x = starting_point
    
    for i in range(iterations):
        gradient = 2*(x+3)
        x = x - learning_rate*gradient
        
        x_values.append(x)
        y_values.append(function(x))
        
        print("Iteration",i,"\n x value is: ",x)
    return x_values,y_values
```

```python
starting_point = 2
learning_rate = 0.1
iterations = 10000
```

```python
x_values, y_values = gradient_descent(starting_point, learning_rate, iterations)
```

```python
x_vals = np.linspace(-10, 4, 100)
y_vals = function(x_vals)
```

```python
plt.plot(x_vals, y_vals, label='y = (x+3)^2')
plt.scatter(x_values, y_values, color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.legend()
plt.show()
```

## Conclusion
Thus, gradient descent algorithm has been implemented. The local minimum for  y=(x+3)Â² was found to be very close to -3 after 161 iterations.
